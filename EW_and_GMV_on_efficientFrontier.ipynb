# Lack of Robustness of the Markowitz Procedure, and the GMW

This show the graph of the equally weighted portfolio and the Global Minimum Variance Portfolio
Both the equally-weight-Portfolio and the Global-Minimum-Variance Portfolio are independent on the estimated Values.
Both Portfolios can avoid the Problem that are caused by the estimated Values.

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import edhec_risk_kit as erk
from matplotlib import pyplot as plt
import numpy as np

'''
Get the Index Returns of years 1996-2000
Generate Expected Returns and Covariance Matrix
'''

ind = erk.get_ind_returns()
er = erk.annualize_returns(ind["1996":"2000"],12)
cov = ind["1996":"2000"].cov()
```

```python
'''
Returns the Max sharped Ratio of Food and Steel by the risk free rate of 10%
'''
l = ["Food", "Steel"]
erk.msr(0.1, er[l], cov.loc[l,l])
```

```python
'''
Returns the expected Returns (annual Returns of Food and Steel)
'''
er[l]
```

```python
'''
The Values is very different even with a small change in values
For Food, instead of 0.114235 use 0.1 and for Steel use 0.13 instead of 0.11580
The Max Sharpe Ratio give a very unrealistic weight. cannot be used.
'''
erk.msr(0.1, np.array([.10, .13]) , cov.loc[l,l])
```

### Let create a graph of Efficient Frontier with Capital Markt line

```python
erk.plot_ef(20, er, cov, show_cml=True, riskfree_rate=0.1)
```
